# Visualize Tensor Notebook

This notebook provides clear demonstrations of all the main functions in the `visualize_tensor` library.

## Overview of Functions

- **`show(tensor)`**: General-purpose tensor visualization that automatically chooses the best display method based on tensor dimensions
- **`show_layers(tensor)`**: Stacked layer visualization for 3D+ tensors, showing each "slice" as a separate layer
- **`show_heatmap(tensor)`**: Heatmap visualization for 2D tensors with color mapping
- **`show_stats(tensor)`**: Statistical summary and distribution plots for tensors
- **`tensor_summary(tensor)`**: Text summary of tensor properties (shape, dtype, range, etc.)

```python
# Import libraries
import numpy as np
import torch
import matplotlib.pyplot as plt

# Import visualize_tensor functions
from visualize_tensor import show, show_layers, show_heatmap, show_stats, tensor_summary

# Check version
import visualize_tensor
print(f"visualize_tensor version: {visualize_tensor.__version__}")
```

```python
# Create sample tensors for testing
print("Creating sample tensors...")

# 1D tensor (vector)
t1d = torch.rand(12)
print(f"1D tensor shape: {t1d.shape}")

# 2D tensor (matrix)
t2d = torch.rand(8, 8)
print(f"2D tensor shape: {t2d.shape}")

# 3D tensor (common for images: height, width, channels)
t3d = torch.rand(28, 28, 3)
print(f"3D tensor shape: {t3d.shape}")

# Large 2D tensor
tbig = torch.randn(32, 64)
print(f"Large 2D tensor shape: {tbig.shape}")

# Large 3D tensor
t3d_big = torch.rand(32, 32, 3)
print(f"Large 3D tensor shape: {t3d_big.shape}")

print("Sample tensors created!")
```

## 1. tensor_summary() - Text Summary

`tensor_summary(tensor)` returns a string with key information about the tensor: shape, data type, value range, and statistics.

```python
print("1D tensor summary:")
print(tensor_summary(t1d))
print("\n2D tensor summary:")
print(tensor_summary(t2d))
print("\n3D tensor summary:")
print(tensor_summary(t3d))
```

## 2. show() - General Tensor Visualization

`show(tensor)` automatically chooses the best visualization method based on tensor dimensions:
- 1D: Line plot
- 2D: Heatmap
- 3D+: Layered visualization

```python
print("▶ show() on 1D tensor (line plot)")
show(t1d, title="1D Tensor - Line Plot")
```

```python
print("▶ show() on 2D tensor (heatmap)")
show(t2d, title="2D Tensor - Heatmap")
```

```python
print("▶ show() on 3D tensor (layered view)")
show(t3d, title="3D Tensor - Layered View")
```

## 3. show_heatmap() - Heatmap Visualization

`show_heatmap(tensor)` creates a color-coded heatmap for 2D tensors. For 3D tensors, it shows separate heatmaps for each channel/layer.

```python
print("▶ show_heatmap() on 2D tensor")
show_heatmap(t2d, title="2D Heatmap", cmap="viridis")
```

```python
print("▶ show_heatmap() on 3D tensor (per-channel)")
show_heatmap(t3d, title="3D Tensor - Per-Channel Heatmaps", cmap="plasma")
```

## 4. show_layers() - Layered 3D Visualization

`show_layers(tensor)` shows 3D+ tensors as stacked "layers" where each 2D slice is displayed as a separate visual layer with depth perspective. Useful for understanding multi-dimensional data like image batches or feature maps.

```python
print("▶ show_layers() on 3D tensor")
show_layers(t3d, title="3D Tensor - Stacked Layers")
```

```python
print("▶ show_layers() with head/tail cropping (for large tensors)")
show_layers(t3d_big, head=5, tail=5, title="Large 3D Tensor - Cropped View")
```

## 5. show_stats() - Statistical Analysis

`show_stats(tensor)` displays comprehensive statistics and distribution plots for tensors, including histograms, box plots, and summary statistics.

```python
print("▶ show_stats() - statistical analysis")
show_stats(tbig, title="Large 2D Tensor Statistics")
```

## Summary

This notebook demonstrated all the main functions in `visualize_tensor`:

1. **`tensor_summary()`** - Quick text overview of tensor properties
2. **`show()`** - Automatic visualization based on dimensions
3. **`show_heatmap()`** - Color-coded heatmaps for 2D/3D data
4. **`show_layers()`** - 3D layered visualization with depth perspective
5. **`show_stats()`** - Statistical analysis and distribution plots

Each function is designed to help you understand your tensors better, whether you're debugging neural networks, exploring data, or just learning PyTorch/NumPy.

### Key Parameters

- `head`/`tail`: Control how many elements to show from start/end of large tensors
- `max_layers`: Limit number of layers shown in 3D visualizations
- `cmap`: Color map for heatmaps (e.g., 'viridis', 'plasma', 'coolwarm')
- `decimals`: Number of decimal places to display
- `title`: Custom title for plots
- `figsize`: Figure size as (width, height) tuple

All functions work with both NumPy arrays and PyTorch tensors automatically!
